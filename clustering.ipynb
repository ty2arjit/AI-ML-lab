```python
import numpy as np
import matplotlib.pyplot as mlt
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
iris=datasets.load_iris()
X=iris.data[:,:2]
kmeans=KMeans(n_clusters=2,random_state=42)
y_kmeans=kmeans.fit_predict(X)
```

```python
mlt.figure(figsize=(10,5))
mlt.scatter(X[:,0],X[:,1],c=y_kmeans,cmap="viridis",edgecolors='k',s=50)
mlt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='red',marker='X',label="Centeroid")
mlt.xlabel('X')
mlt.ylabel('Y')
mlt.title('K-Means Clustering on Iris Dataset')
mlt.legend()
mlt.show()
```
