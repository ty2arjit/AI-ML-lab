```python
import pandas as pd
import numpy as np
df=pd.read_csv('city_population_vs_profit.csv').to_numpy()
ratio=0.3  
rows=df.shape[0]
train_size=int(rows*ratio)
train=df[0:train_size]
test=df[train_size:]
print("Train dataframe: ",train)
print("Test dataframe: ",test)
def cost(x,y,w,b):
    total_cost=0
    for i in range(rows):
        temp=w*x[i]+b
        total_cost+=(temp-y[i])**2
    return total_cost/(2*rows)
x=df['Population of city ($1000)']
y=df['Profit of restaurants ($10K)']
total_cost=cost(x,y,0,0)
print(total_cost)
def compute_gradient_w(x,y,w,b):
    sum=0.0
    for i in range(rows):
        sum+=(w*x[i]+b-y[i])*x[i]
    return sum/rows
def compute_gradient_b(x,y,w,b):
    sum=0.0
    for i in range(rows):
        sum+=(w*x[i]+b-y[i])
    return sum/rows
def gradient_descent(x,y,w,b,alpha,iterations):
    for i in range(iterations):
        dj_dw=compute_gradient_w(x,y,w,b)
        dj_db=compute_gradient_b(x,y,w,b)
        w-=alpha*dj_dw
        b-=alpha*dj_db
    return w,b
#w,b=gradient_descent(train[0],train[1],0,0,0.01,100)
x1=train[:,0]
y1=train[:,1]
w,b=gradient_descent(x1,y1,0,0,0.01,100)
print(w,b)
y2=[]
for i in x:
    y2.append(w*i+b)
import matplotlib.pyplot as mlp
mlp.plot(x,y2)
mlp.scatter(x,y)
mlp.show()
```
